```python
import matplotlib.pyplot as plt

# Matrix sizes
matrix_sizes = [200, 300, 900, 1300]

# Performance values
python_performance = [python_200, python_300, python_900, python_1300]
cpp_normal_performance = [cpp_normal_200, cpp_normal_300, cpp_normal_900, cpp_normal_1300]
cpp_parale_performance = [cpp_parale_200, cpp_parale_300, cpp_parale_900, cpp_parale_1300]

# Plotting
plt.figure(figsize=(12, 6))

plt.plot(matrix_sizes, python_performance, label='Python', marker='o')
plt.plot(matrix_sizes, cpp_normal_performance, label='C++ Normal', marker='o')
plt.plot(matrix_sizes, cpp_parale_performance, label='C++ Paralelo', marker='o')

plt.xlabel('Tamanho da Matriz')
plt.ylabel('Tempo de Execu√ß√£o (s)')
plt.title('Compara√ß√£o de Performance entre Python e C++')
plt.legend()
plt.grid(True)

plt.show()
```
