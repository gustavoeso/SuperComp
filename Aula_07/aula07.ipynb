As análises mostram uma disparidade significativa entre o tempo de execução das heurísticas aleatórias e da busca exaustiva, além de uma diferença notável na qualidade das soluções encontradas.

### Tempo de Execução:
As heurísticas aleatórias (tanto "Embaralhar e Preencher" quanto "Seleção Baseada em Probabilidade") são extremamente rápidas em comparação com a busca exaustiva. Os tempos de execução das heurísticas são da ordem de microssegundos (10⁻⁶ segundos), enquanto a busca exaustiva leva de segundos a minutos para processar as mesmas entradas. Essa diferença é tão grande que os tempos das heurísticas não aparecem de forma significativa nos gráficos quando comparados ao tempo da busca exaustiva.

Essa rapidez das heurísticas ocorre porque elas seguem abordagens simplificadas e não exploram todas as combinações possíveis, enquanto a busca exaustiva analisa todas as opções para encontrar a solução ótima.

### Qualidade das Soluções (Valores Obtidos e Pesos das Mochilas):
No entanto, essa velocidade das heurísticas tem um custo significativo: a qualidade das soluções encontradas é muito inferior à obtida pela busca exaustiva.

- **Entrada 1**: A busca exaustiva encontrou uma solução com um valor de 601, enquanto as heurísticas produziram valores médios de 299.8 (Embaralhar e Preencher) e 339.4 (Seleção Baseada em Probabilidade). Isso representa uma perda considerável na qualidade da solução, com as heurísticas conseguindo aproximadamente metade do valor encontrado pela busca exaustiva.

- **Entrada 2**: Aqui, a busca exaustiva encontrou um valor de 870, enquanto as heurísticas produziram valores médios de 242.2 e 226.2. A diferença é ainda mais pronunciada, com as heurísticas conseguindo apenas cerca de 25% do valor ótimo.

- **Entrada 4**: O valor encontrado pela busca exaustiva foi 648, comparado aos valores médios das heurísticas de 499.4 e 404.6. Novamente, as heurísticas ficam bem aquém do valor ótimo.

### Conclusão:
Embora as heurísticas aleatórias ofereçam uma solução extremamente rápida, essa eficiência vem ao custo de uma perda significativa na qualidade da solução. As heurísticas falham em explorar o espaço de solução adequadamente, resultando em valores muito abaixo do ótimo e em mochilas que não são preenchidas de forma eficiente em termos de valor. Isso ilustra o compromisso clássico entre tempo de execução e qualidade da solução: enquanto a busca exaustiva garante a solução ótima, ela requer um tempo de execução muito maior, o que pode ser impraticável para problemas de grande escala. Por outro lado, as heurísticas, embora rápidas, podem produzir soluções "extremamente erradas" quando comparadas ao ideal.
