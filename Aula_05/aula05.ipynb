```python
########### 15 iteracoes ###########
sem_otimizacoes_15 = 0.843943
otimizacao_basica_15 = 0.2248
otimizacao_padrao_15 = 0.17555
otimizacao_avancada_15 = 0.0790481

print("sem otimizacao 15: ", sem_otimizacoes_15)
print("otimizacao basica 15: ", otimizacao_basica_15)
print("otimizacao padrao 15: ", otimizacao_padrao_15)
print("otimizacao avancada 15: ", otimizacao_avancada_15)

########### 30 iteracoes ###########
sem_otimizacoes_30 = 1.69491
otimizacao_basica_30 = 0.453122
otimizacao_padrao_30 = 0.350733
otimizacao_avancada_30 = 0.158565

print("sem otimizacao 30: ", sem_otimizacoes_30)
print("otimizacao basica 30: ", otimizacao_basica_30)
print("otimizacao padrao 30: ", otimizacao_padrao_30)
print("otimizacao avancada 30: ", otimizacao_avancada_30)
########### 50 iteracoes ###########
sem_otimizacoes_50 = 2.80785
otimizacao_basica_50 = 0.757978
otimizacao_padrao_50 = 0.581122
otimizacao_avancada_50 = 0.257782

print("sem otimizacao 50: ", sem_otimizacoes_50)
print("otimizacao basica 50: ", otimizacao_basica_50)
print("otimizacao padrao 50: ", otimizacao_padrao_50)
print("otimizacao avancada 50: ", otimizacao_avancada_50)

########### 100 iteracoes ###########
sem_otimizacoes_100 = 5.62783
otimizacao_basica_100 = 1.50629
otimizacao_padrao_100 = 1.1654
otimizacao_avancada_100 = 0.51562

print("sem otimizacao 100: ", sem_otimizacoes_100)
print("otimizacao basica 100: ", otimizacao_basica_100)
print("otimizacao padrao 100: ", otimizacao_padrao_100)
print("otimizacao avancada 100: ", otimizacao_avancada_100)
```

```python
import matplotlib.pyplot as plt

# Matrix sizes
matrix_sizes = [15, 30, 50, 100]

# Performance values
sem_otimizacoes = [sem_otimizacoes_15, sem_otimizacoes_30, sem_otimizacoes_50, sem_otimizacoes_100]
otimizacao_basica = [otimizacao_basica_15, otimizacao_basica_30, otimizacao_basica_50, otimizacao_basica_100]
otimizacao_padrao = [otimizacao_padrao_15, otimizacao_padrao_30, otimizacao_padrao_50, otimizacao_padrao_100]
otimizacao_avancada = [otimizacao_avancada_15, otimizacao_avancada_30, otimizacao_avancada_50, otimizacao_avancada_100]

# Plotting
plt.figure(figsize=(12, 6))

plt.plot(matrix_sizes, sem_otimizacoes, label='Sem Otimizações', marker='o')
plt.plot(matrix_sizes, otimizacao_basica, label='Otimização Básica', marker='o')
plt.plot(matrix_sizes, otimizacao_padrao, label='Otimização Padrão', marker='o')
plt.plot(matrix_sizes, otimizacao_avancada, label='Otimização Avançada', marker='o')

plt.xlabel('Número de Iterações')
plt.ylabel('Tempo de Execução (s)')
plt.title('Comparação de Performance com Diferentes Níveis de Otimização')
plt.legend()
plt.grid(True)

plt.show()
```

**Impacto das Diferentes Flags de Otimização:**

1. **Sem Otimização:**
   - O tempo de execução cresce significativamente com o número de iterações, indicando que o processo não está sendo otimizado de forma alguma. Para 15 iterações, o tempo foi de 0.843943 segundos, enquanto para 100 iterações, saltou para 5.62783 segundos. Isso mostra que o aumento no número de iterações impacta diretamente o tempo de execução sem qualquer mitigação.

2. **Otimização Básica:**
   - A otimização básica já traz uma redução substancial no tempo de execução em relação à versão sem otimização. Por exemplo, o tempo para 100 iterações foi reduzido para 1.50629 segundos, em comparação com os 5.62783 segundos sem otimização. Essa flag de otimização provavelmente inclui melhorias básicas, como eliminação de redundâncias e otimização de loops, que reduzem o tempo de execução de forma significativa, mas ainda há margem para melhorias.

3. **Otimização Padrão:**
   - A otimização padrão vai além da básica, mostrando uma redução ainda maior no tempo de execução. Com 100 iterações, o tempo caiu para 1.1654 segundos. Isso sugere que essa flag de otimização realiza otimizações adicionais, como a reordenação de operações, inlining de funções, ou outras técnicas que aumentam a eficiência do código.

4. **Otimização Avançada:**
   - A otimização avançada oferece os maiores ganhos, com o tempo de execução caindo para 0.51562 segundos para 100 iterações. Este nível de otimização provavelmente envolve técnicas mais agressivas, como vetorização, paralelização, e outras transformações de alto nível, que resultam em um desempenho muito superior. A redução no tempo de execução é a mais significativa neste caso, especialmente em situações com um número elevado de iterações.


## Usando o Gprofile

### Análise do Flat Profile

No *Flat Profile*, temos uma visão direta do tempo gasto por cada função isoladamente. Aqui estão as principais observações:

1. **`apply_filter`**:
   - **Tempo gasto**: 97,73% do tempo total (0,86 segundos).
   - **Número de chamadas**: 100.000.000.
   - **Tempo por chamada**: 8,60 nanosegundos por chamada.
   - **Conclusão**: Esta função é a principal consumidora de tempo do programa, representando quase todo o tempo de execução.

2. **`main`**:
   - **Tempo gasto**: 2,27% do tempo total (0,02 segundos).
   - **Conclusão**: A função `main` basicamente invoca a função `apply_filter` e não contribui significativamente para o tempo total, além do tempo gasto dentro dela própria.

### Análise do Call Graph

O *Call Graph* fornece uma visão das relações entre as funções, incluindo quem chama quem e o tempo propagado entre as funções.

1. **`main`**:
   - **Tempo total (self + children)**: 100% (0,88 segundos).
   - **Tempo dentro da função (`self`)**: 0,02 segundos.
   - **Tempo propagado para a função `apply_filter` (`children`)**: 0,86 segundos.
   - **Número de chamadas**: `main` chama `apply_filter` 100.000.000 vezes.

2. **`apply_filter`**:
   - **Tempo total (self)**: 0,86 segundos.
   - **Número de chamadas**: 100.000.000 vezes.
   - **Conclusão**: A função `apply_filter` é chamada repetidamente pela `main`, consumindo a maior parte do tempo de execução.

### Funções Críticas

A função crítica identificada é a **`apply_filter`**, que consome 97,73% do tempo total de execução do programa. Essa função é claramente o gargalo de desempenho.
