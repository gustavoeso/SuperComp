- **Qualidade da Solução**: A busca exaustiva continua sendo a abordagem que oferece a melhor qualidade de solução, mas seu custo computacional elevado a torna impraticável para grandes problemas. A "Mochila Cheia" e a heurística aleatória falharam em encontrar soluções de alta qualidade, especialmente para entradas mais complexas.

- **Tempo de Execução**: Todas as abordagens de busca local apresentaram tempos de execução extremamente rápidos, tornando-as boas opções quando a solução ótima não é uma exigência e o tempo de processamento é uma prioridade. Mesmo o Hill Climbing, que obteve as melhores soluções entre os algoritmos locais, conseguiu tempos de execução muito rápidos.

- **Ganho com Heurísticas Locais**: Embora as heurísticas locais não encontrem a solução ótima, elas oferecem uma boa relação entre tempo e qualidade da solução. Em problemas maiores, a diferença de tempo entre a busca exaustiva e essas heurísticas será ainda mais significativa.

### Avaliação das Estratégias:

- **Hill Climbing** é uma boa escolha quando se busca uma solução razoavelmente boa em tempo muito curto, pois tende a encontrar boas soluções sem custo computacional elevado.
- **Substituição de Objeto** tem potencial para encontrar boas soluções, mas pode ser inconsistente em seus resultados, principalmente devido à natureza da solução inicial.
- **Mochila Cheia** é rápida, mas não oferece soluções de alta qualidade. No entanto, ela pode ser útil em contextos onde é necessário testar rapidamente várias soluções e melhorar a partir delas.

Esses resultados sugerem que, para problemas de escalas maiores ou com restrições de tempo, heurísticas de busca local como Hill Climbing oferecem uma boa alternativa à busca exaustiva, enquanto ainda conseguem obter soluções de qualidade aceitável.
